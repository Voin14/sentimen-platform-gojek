```python
from google_play_scraper import app
import pandas as pd
import numpy as np
```

```python
from google_play_scraper import Sort, reviews
result, continuation_token = reviews(
    'com.gojek.app',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=15000,
    filter_score_with=None,
)
```

```python
gojek = pd.DataFrame(np.array(result), columns=['review'])
gojek = gojek.join(pd.DataFrame(gojek.pop('review').tolist()))
```
